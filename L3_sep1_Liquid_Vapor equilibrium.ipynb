# Liquid/Vapor Equilibrium : relative volatility

At 373 K, the vapor pressure of n-heptane is 106 kN/m2  and the one of toluene is 73,7 kN/m2

**Calculate the different molar fraction in the gas and liquid phase at equilibrium for a total pressure of 101,3 kN/m2**

>The pressure of the mixture is the sum of the partial pressure :
>$$P_A+P_B=P$$
>The writing of the Raoult law for ideal components leads to :
>$$x_A P_A^0+(1-x_A)P_B^0=P$$
The molar fraction of A can be calculated from :
>$$x_A=\frac{P-P_B^0}{P_A^0-P_B^0}$$


```python
P=101.3 #kN/m2
PA0=106 #kN/m2 the n-heptane is the more volatile compound
PB0=73.7 #kN/m2
xA=(P-PB0)/(PA0-PB0)
yA=xA*PA0/P
yB=1-yA
xB=1-xA
print ('yA  =', round(yA,3),'     yB  =', round(yB,3))
print ('xA  =', round(xA,3),'     xB  =', round(xB,3))
```

A une température donnée on peut déterminer la pression de vapeur du binaire en fonction de la composition.

```python
plt.plot([0,1],[0,PA0],'y',label='Partial pressure of A')
plt.plot([0,1],[PB0,0],'b',label='Partial pressure of B')        
plt.plot([0,1],[PB0,PA0],'g',label='Vapor pressure of the mixture')
plt.text(0.4,20,'T=373 K',size='15')
plt.xlabel('$x_{BZ}$')
plt.ylabel('Pressure in $kN.m^{-2}$')
plt.legend()
plt.grid()
plt.show()
```

**Calculate the relative volatility.**
>The relative volatility is defined as :
>$$\alpha=\frac{y_A x_B}{x_A y_B}=\frac{P_A^0}{P_B^0}$$

```python
import numpy as np
import matplotlib.pyplot as plt
alpha1=yA*xB/(xA*yB)
alpha2=PA0/PB0
print ('The relative volatility is', round(alpha1,3), ' or ', round(alpha2,3))
```

La volatilité relative, $\alpha$ peut être utilisée pour tracer le diagramme carré $y(x)$. 

$$y=\frac{\alpha x}{1+(\alpha-1)x}$$

On fait alors l'hypothèse qu'$\alpha$ est constant alors qu'$\alpha$ peut varier selon la température (vous verrez dans l'exercice suivant que cette variation est quelquefois négligeable).

```python
#plot of square diagram y=f(x)
x=np.linspace(0,1,101)
y=alpha1*x/(1+(alpha1-1)*x)
yb=x
plt.plot(x,y)
plt.plot(x,yb,'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=f(x) plot with the assumption of a constant relative volatility')
plt.text(0.3,0.2,'P={} kN.m-2'.format(P),size='15')
plt.show()
```
